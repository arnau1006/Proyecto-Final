```python
ca_completa = nula.merge(
    Calidad_Aire,
    on = ['Año', 'Emisión', 'Emisión_ID', 'Provincia', 'Provincia_ID'],
    how = 'left'
)
ca_completa
```

```python
df = pd.DataFrame(ca_completa)

def fill_missing_values(df):
    # Crear una copia del DataFrame para preservar los valores originales
    df_copy = df.copy()

    # Interpolación lineal y relleno de valores nulos por grupo de 'Emisión' y 'Provincia'
    df_copy['Interpolacion'] = df_copy.groupby(['Emisión', 'Provincia'])['Emisión Anual'].transform(lambda group: group.interpolate(method='linear'))

    # Rellenar los valores nulos restantes con el promedio del grupo
    df_copy['Interpolacion'] = df_copy.groupby(['Emisión', 'Provincia'])['Interpolacion'].transform(lambda group: group.fillna(group.mean()))

    # Conservar los valores originales y reemplazar solo los nulos en el DataFrame original
    df['Emisión Anual'] = df['Emisión Anual'].combine_first(df_copy['Interpolacion'])

    return df

# Aplicar la función para rellenar los valores faltantes
Calidad_Aire_Completa = fill_missing_values(df)
```
