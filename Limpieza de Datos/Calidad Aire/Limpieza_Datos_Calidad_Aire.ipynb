```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as snf
import seaborn as sns
```

```python
data_as = pd.read_csv("As_DD_2012.csv", delimiter=";")
data_bap = pd.read_csv("BaP_DD_2012.csv", delimiter=";")
data_cd = pd.read_csv("Cd_DD_2012.csv", delimiter=";")
data_ni = pd.read_csv("Ni_DD_2012.csv", delimiter=";")
data_pb = pd.read_csv("Pb_DD_2012.csv", delimiter=";")
data_pm10 = pd.read_csv("PM10_DD_2012.csv", delimiter=";")
data_pm25 = pd.read_csv("PM2.5_DD_2012.csv", delimiter=";")
```

```python
data_as["EMISION"] = "As"
data_bap["EMISION"] = "BaP"
data_cd["EMISION"] = "Cd"
data_ni["EMISION"] = "Ni"
data_pb["EMISION"] = "Pb"
data_pm10["EMISION"] = "PM10"
data_pm25["EMISION"] = "PM25"
```

```python
all_data = pd.concat([data_as, data_bap, data_cd, data_ni, data_pb, data_pm10, data_pm25], ignore_index=True)
```

```python
day_columns = [f'D{i:02}' for i in range(1, 32)]
all_data[day_columns] = all_data[day_columns].interpolate(axis=1, method='linear', limit_direction='both')
```

```python
medias = []
for i in range(len(all_data)):
    row = all_data.loc[i, day_columns]
    if row.count() >= 15:
        valor_media = row.mean()
    else:
        valor_media = np.nan
    medias.append(valor_media)
```

```python
all_data["MEDIA MENSUAL"] = medias
```

```python
result_12 = all_data.groupby(["EMISION", "PROVINCIA", "ANNO"]).agg(
    MEDIAANUAL_EMISION=pd.NamedAgg(column="MEDIA MENSUAL", aggfunc="mean")
).reset_index()
```

```python
result_12.MEDIAANUAL_EMISION = result_12.MEDIAANUAL_EMISION.round(4)
```

```python
result_12
```
